```python
import pandas as pd

df = pd.read_excel("../data/base.xlsx")
```

```python
import pandas as pd

tmp = df[['position_name', 'gender']].copy()

tmp = tmp.dropna(subset=['position_name'])

tmp['position_name'] = tmp['position_name'].str.split(r'[,;|]')

tmp = tmp.explode('position_name').reset_index(drop=True)

tmp['position_name'] = tmp['position_name'].str.strip()

tmp = tmp[tmp['position_name'] != '']

tmp['gender'] = tmp['gender'].fillna('Пропуск')

pivot = pd.crosstab(tmp['position_name'], tmp['gender'])

for col in ['Мужской', 'Женский', 'Пропуск']:
    if col not in pivot.columns:
        pivot[col] = 0

pivot = pivot[['Мужской', 'Женский', 'Пропуск']]

pivot['Всего'] = pivot.sum(axis=1)

pivot['Процент мужчин'] = (
    pivot['Мужской'] / (pivot['Мужской'] + pivot['Женский']) * 100
).round(2)

pivot = pivot.sort_values('Всего', ascending=False)

pivot.to_excel("../data/positions_and_gender.xlsx")
```
