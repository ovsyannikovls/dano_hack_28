Файл для созданния таблицы для финальной обработки

```python
import pandas as pd

predf = pd.read_excel("../../data/interim/preclean_base.xlsx")
```

Код для создания очищенной таблицы V1.0 (параметры: salary, gender, experience, education_type, region_code, industry_code)

Дополнительные фильтры:

- birthday_mistake == 0/"";
- busy_type == "Полная занятость";
- country == "Российская Федерация";
- inner_info_status == "Одобрено";
- experience_mistake == 0;
- inactive == 1;
- relocation == 0/""; (так как достаточно много пропусков, при большом кол-ве 0)
- retraining_capability == 1/""; (так как достаточно много пропусков, при большом кол-ве 1)
- business_trips == 0;
- gender != "";
- salary > 1000;

```python
df = predf.copy()

mask = (
    (df["birthday_mistake"].isna() | (df["birthday_mistake"] == 0)) &
    (df["busy_type"] == "Полная занятость") &
    (df["country"] == "Российская Федерация") &
    (df["inner_info_status"] == "Одобрено") &
    (df["experience_mistake"] == 0) &
    (df["inactive"] == 1) &
    (df["relocation"].isna() | (df["relocation"] == 0)) &
    (df["retraining_capability"].isna() | (df["retraining_capability"] == 1)) &
    (df["business_trips"] == 0) &
    (df["gender"].notna()) &
    (df["gender"] != "") &
    (df["salary"] > 1000)
)

final_cols = [
    "salary",
    "gender",
    "experience",
    "education_type",
    "region_code",
    "industry_code"
]

processed_df = df.loc[mask, final_cols].copy()

processed_df["education_type"] = processed_df["education_type"].fillna("Не указано")
processed_df["industry_code"] = processed_df["industry_code"].fillna("None")

processed_df = processed_df.dropna(
    subset=["salary", "gender", "experience", "region_code"]
).reset_index(drop=True)

# Проверка данных

print(processed_df.head())
print(len(processed_df))
```

```python
processed_df.to_excel("../../data/processed/v10_clean_base.xlsx", index=False)
```
