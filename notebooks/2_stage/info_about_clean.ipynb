Анализ датасета

```python
import pandas as pd

df = pd.read_excel("../../data/interim/preclean_base.xlsx")
```

Кол-во регионов и отраслей

```python
print("Число регионов:", df["region_code"].nunique())
print("Число отраслей:", df["industry_code"].nunique())
```

Пропуски по столбцам (важно)

```python
missing_by_col = pd.DataFrame({
    "missing_count": df.isna().sum(),
    "missing_pct": (df.isna().mean() * 100).round(2)
}).sort_values("missing_count", ascending=False)

missing_by_col.to_excel("../../data/interim/missing_by_col.xlsx")
```

Распределение гендеров по регионам

```python

gender_by_region = pd.crosstab(df["region_code"], df["gender"])

print(f"Всего отмеченных женщин: {df["gender"].eq("Женский").sum()}")
      
print(f"Всего отмеченных мужчин: {df["gender"].eq("Мужской").sum()}")

gender_by_region.to_excel("../../data/interim/gender_by_region.xlsx")
```

Зарплата по регионам

```python
salary_by_region = (
    df.groupby("region_code")["salary"]
    .agg(n="count", mean_salary="mean", median_salary="median")
    .reset_index()
)

salary_by_region.to_excel("../../data/interim/salary_by_region.xlsx")
```

Опыт по регионам

```python
experience_by_region = (
    df.groupby("region_code")["experience"]
    .agg(n="count", mean_experience="mean", median_experience="median")
    .reset_index()
)

experience_by_region.to_excel("../../data/interim/experience_by_region.xlsx")
print(experience_by_region)
```

Средняя заполняемость объявления по регионам

```python
fullness_by_region = (
    df.groupby("region_code")["inner_info_fullness_rate"]
    .agg(n="count", mean_fullness="mean", median_fullness="median")
    .reset_index()
)
print(fullness_by_region)
```

Зарплата по полу внутри региона

```python

salary_gender_region = (
    df.groupby(["region_code", "gender"])["salary"]
    .agg(n="count", mean_salary="mean", median_salary="median")
    .reset_index()
)
print(salary_gender_region)
salary_gender_region.to_excel("../../data/interim/salary_gender_region.xlsx")
```
