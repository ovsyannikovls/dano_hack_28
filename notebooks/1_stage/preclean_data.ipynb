```python
import pandas as pd

df = pd.read_excel("../../data/raw/base.xlsx")
```

Функция для замены нулевых значений и задавание формата столбцов

```python
def cast_columns(df: pd.DataFrame) -> pd.DataFrame:
    df = df.copy()

    df = df.replace(
        [r"^\s*$", "NaN", "nan", "None", "null", "NULL", "<NA>", "N/A", "n/a"],
        pd.NA,
        regex=True,
    )

    df['region_code'] = (
        df['region_code']
        .astype(str)
        .str.replace(r'\.0$', '', regex=True)
        .str.replace(r'\D+', '', regex=True)
        .str.zfill(13)
        .str[:2]
    )

    string_cols = [
        "gender",
        "country",
        "skills",
        "additional_skills",
        "education_type",
        "other_info",
        "industry_code",
        "busy_type",
        "position_name",
        "schedule_type",
        "inner_info_status",
        "profession_code",
        "worldskills_inspection_status",
        "worldskills_international_name",
        "worldskills_russian_name",
        "worldskills_skill_abbreviation",
        "worldskills_type",
        "add_certificates",
        "nark_certificate",
        "drive_licences",
        "locality",
        "id_candidate",
        "id_cv",
        "id_user_inner_info",
    ]

    int_cols = [
        "experience",
        "salary",
        "inner_info_fullness_rate",
        "region_code",
        "birthday",
    ]

    bool_cols = [
        "business_trips",
        "relocation",
        "retraining_capability",
        "worldskills_is_international",
        "inactive",
        "birthday_mistake",
        "experience_mistake",
        "driver_licence_a",
        "driver_licence_b",
        "driver_licence_c",
        "driver_licence_d",
        "driver_licence_e",
        "schedule_type_1",
        "schedule_type_2",
        "schedule_type_3",
        "schedule_type_4",
        "schedule_type_5",
        "schedule_type_6",
    ]

    date_cols = [
        "date_creation",
        "date_inactivation",
        "date_last_updated",
        "date_modify_inner_info",
        "date_publish",
    ]

    datetime_cols = [
        "date_time_publish",
    ]

    time_cols = [
        "time_publish",
    ]

    for col in string_cols:
        if col in df.columns:
            df[col] = df[col].astype("string").str.strip()

    for col in int_cols:
        if col in df.columns:
            df[col] = pd.to_numeric(df[col], errors="coerce").astype("Int64")

    for col in bool_cols:
        if col in df.columns:
            s = pd.to_numeric(df[col], errors="coerce")
            s = s.where(s.isin([0, 1]) | s.isna())
            df[col] = s.astype("Int64")

    for col in date_cols:
        if col in df.columns:
            df[col] = pd.to_datetime(df[col], errors="coerce").dt.normalize()

    for col in datetime_cols:
        if col in df.columns:
            df[col] = pd.to_datetime(df[col], errors="coerce")

    for col in time_cols:
        if col in df.columns:
            df[col] = pd.to_datetime(
                df[col],
                format="%H:%M:%S",
                errors="coerce"
            ).dt.time

    return df


df = cast_columns(df)

# Проверка форматов столбцов

print(df.dtypes)
```

```python
df.to_excel("../../data/interim/preclean_base.xlsx", index=False)
```
